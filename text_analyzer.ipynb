# Creating function for count analysis

```python
# creating function for count analysis for different types of tokens 
def analyse_token_count_statistically_and_visually(text, token_list, token_type, title):
  
  print('Count of', token_type,': \n')

  #initilaizing dictionary with each token and its count as 0
  token_dict = dict(zip(token_list,[0]*len(token_list)))

  #updating total count of each token
  for token in token_list:
    token_dict[token] = text.count(token)

  #sorting tokens  from maximum to minimum count in dictionary
  token_dict= dict(sorted(token_dict.items(), key=lambda item: item[1])[::-1])

  print(token_dict,"\n")

  #plotting ppie plot by providinng size, label and percentage of count
  plt.figure(figsize=(16, 10))
  plt.pie(token_dict.values(),labels = token_dict.keys(), autopct='%.3f%%')

  plt.title(title)
  Artist.set_visible(fig, True)
```

# Calling count analysis function for lowercase letters

```python
#parsing wikipedia text data 
text = wikipedia.page('Indian_Premier_League').content
#creating token list to be analyzed
token_list = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
token_type = 'lowercase letters'
title = 'Percentage count of each lowercase letter'


analyse_token_count_statistically_and_visually(text, token_list, token_type, title)
```

# Analyzing number of seasons played table

```python
wiki_url = 'https://en.wikipedia.org/wiki/Indian_Premier_League'
class_val="wikitable plainrowheaders"

response = requests.get(wiki_url)
soup= BeautifulSoup(response.text, 'html.parser')
#using findAll to search all present tables with that name, using indexing to use target table
class_val="wikitable"
teams_performance_table = soup.findAll('table',class_=class_val)
teams_performance_table = teams_performance_table[2]
df = pd.read_html(str(teams_performance_table))[0]
df
```

```python
seasons_played=[]
team = []
shortname = []
#creating shortnames of teams by taking their first character of their consituent name portions
for s_played in df['No. of seasons played']:
  seasons_played.append(int(s_played))

for _team in df['Team.1']:
  team.append(_team)
  short = ""
  tokens= _team.split()
  for i in tokens:
    short = short + i[0]
  shortname.append(short)


seasons_played_df = pd.DataFrame({'Team':team,'Seasons Played':seasons_played,'shortname':shortname})
seasons_played_df
```

```python
fig  = plt.figure(figsize=(16, 10))
plt.bar(seasons_played_df['shortname'],seasons_played_df['Seasons Played'])
plt.title('Number of seasons played by each Team')
Artist.set_visible(fig, True)
```
